```python
from colorama import Fore, Style, init

# Initialize Colorama
init()

def colorize_data(data, parent_key=None):
    if isinstance(data, dict):
        print('{')
        for i, (k, v) in enumerate(data.items()):
            print(f"  {repr(k)}: ", end='')
            colorize_data(v, k)  # Pass the dictionary key as parent_key
            if i < len(data) - 1:
                print(',')
        print('}')
    elif isinstance(data, list):
        print('[', end='')
        for i, item in enumerate(data):
            colorize_data(item)  # No specific key for list items
            if i < len(data) - 1:
                print(', ', end='')
        print(']', end='')
    elif isinstance(data, str):
        # Check if we're processing the 'text' key, and that we have a parent reference
        if parent_key == 'text' and parent_ref and parent_ref.get('_meta', {}).get('title') == 'negative':
            print(Fore.RED + repr(data) + Style.RESET_ALL, end='')
        elif parent_key == 'text' and parent_ref and parent_ref.get('_meta', {}).get('title') == 'positive':
            print(Fore.GREEN + repr(data) + Style.RESET_ALL, end='')
        else:
            print(repr(data), end='')
    else:
        print(repr(data), end='')

# Make sure to modify calls to colorize_data to include the new parameter:
def colorize_print_recursive(data, parent=None):
    global parent_ref
    parent_ref = parent

    if isinstance(data, dict) and '_meta' in data:
        print('{', end='')
        for i, (key, value) in enumerate(data.items()):
            if key == '_meta' or key == 'inputs':
                print(f"  {repr(key)}: ", end='')
                colorize_print_recursive(value, data if key == 'inputs' else None)
                if i < len(data) - 1:
                    print(',', end='')
        print('}', end='')

    elif isinstance(data, dict):
        colorize_data(data)  # No key when calling from a dictionary
    elif isinstance(data, list):
        colorize_data(data)  # No key when calling from a list
    else:
        colorize_data(data)  # No key for non-dict/list data

def print_colorize(data):
    for key in data:
        print(f"{repr(key)}: ", end='')
        colorize_print_recursive(data[key])
        print()  # Newline for each top-level item

# Your provided data
data = {
    "6": {
        "inputs": {
            "text": "positive",
            "clip": ["4", 1]
        },
        "class_type": "CLIPTextEncode",
        "_meta": {
            "title": "positive"
        }
    },
    "7": {
        "inputs": {
            "text": "negative",
            "clip": ["4", 1]
        },
        "class_type": "CLIPTextEncode",
        "_meta": {
            "title": "negative"
        }
    }
}

# Print using our new method
print_colorize(data)
```

```python
from colorama import Fore, Style, init

# Initialize colorama
init(autoreset=True)

# Your data
data = {
    "6": {
        "inputs": {
            "text": "positive",
            "clip": ["4", 1]
        },
        "class_type": "CLIPTextEncode",
        "_meta": {
            "title": "positive"
        }
    },
    "7": {
        "inputs": {
            "text": "negative",
            "clip": ["4", 1]
        },
        "class_type": "CLIPTextEncode",
        "_meta": {
            "title": "negative"
        }
    }
}

def print_colored_text(data):
    for key, value in data.items():
        text = value['inputs']['text']
        title = value['_meta']['title']
        
        if title.lower() == 'positive':
            print(Fore.GREEN + text)
        elif title.lower() == 'negative':
            print(Fore.RED + text)
        else:
            print(text)

# Call the function
print_colored_text(data)
```
